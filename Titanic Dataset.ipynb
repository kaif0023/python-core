```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
data =sns.load_dataset("titanic")
data
```

```python
for i in data.columns:
    print(i,"--------",data[i].isnull().sum())
```

```python
for i in data.columns:
    if ((data[i].isnull().sum()*100/data.shape[0])>=70):
        data.drop(columns=[i],inplace=True)
```

```python
data.info()
```

```python
data["age"].fillna(data["age"].median(),inplace=True)
```

```python
data.info()
```

```python
data.drop(columns=data[["survived","pclass","adult_male","embarked","who"]],inplace=True)
```

```python
data.info()
```

```python
data
```

```python
data["embark_town"].fillna(data["embark_town"].mode()[0],inplace=True)
```
